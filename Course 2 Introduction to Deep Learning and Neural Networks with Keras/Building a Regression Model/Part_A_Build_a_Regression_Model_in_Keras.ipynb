## Import the necessary libraries

```python
import keras

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import warnings
warnings.simplefilter('ignore', FutureWarning)
```

## Let's download the data and read it into a pandas dataframe.

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

# Build the neural network

```python
# Separate features (X) and target variable (y)
concrete_data_columns = concrete_data.columns
X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]  # Assuming 'Strength' is the target
y = concrete_data['Strength']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dense(1))

    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

# Train and test the network

```python
# build the model
model = regression_model()
```

```python
# fit the model
model.fit(X_train, y_train, epochs=50, batch_size=32)
```

## Calculate mean-squared error of the model

```python
# Evaluate the model on the test data
predictions = model.predict(X_test)

# Compute the mean squared error
mse = mean_squared_error(y_test, predictions)

print("Mean Squared Error:", mse)
```

# Do the above steps 50 times creating a list of 50 mean squares

```python
# define function to train and evaluate the model using mean-squared error
def train_and_evaluate(model, X_train, y_train, X_test, y_test, epochs=50):
  model.fit(X_train, y_train, epochs=epochs, batch_size=32)
  predictions = model.predict(X_test)
  mse = mean_squared_error(y_test, predictions)
  return mse

# Create a list to store mean squared errors
mse_list = []

# Repeat the training and evaluation process 50 times
for _ in range(50):
  model = Sequential()
  model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
  model.add(Dense(1))

  model.compile(optimizer='adam', loss='mean_squared_error')


  mse = train_and_evaluate(model, X_train, y_train, X_test, y_test)
  mse_list.append(mse)

# Calculate mean and standard deviation of MSE
mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)

print("MSE List:",mse_list)
print("Mean MSE:", mean_mse)
print("Standard Deviation of MSE:", std_mse)
```
